```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Weather.csv")

print("CSV Loaded Successfully!\n")
print("Available Columns:\n")
print(list(df.columns))
```

```python
x_col = input("Enter the column name for X axis: ")
y_col = input("Enter the column name for Y axis: ")
```

```python
print("\nChoose Chart Type:")
print("1. Line Chart")
print("2. Bar Chart")

chart_choice = input("\nEnter your choice (1/2): ")
```

```python
plt.figure()

if chart_choice == "1":
    plt.plot(df[x_col], df[y_col])
    plt.title(f"Line Chart - {y_col} vs {x_col}")
elif chart_choice == "2":
    plt.bar(df[x_col], df[y_col])
    plt.title(f"Bar Chart - {y_col} vs {x_col}")
else:
    print("Invalid choice!")
    raise SystemExit

plt.xlabel(x_col)
plt.ylabel(y_col)
plt.grid(True)
plt.tight_layout()
plt.show()
```
